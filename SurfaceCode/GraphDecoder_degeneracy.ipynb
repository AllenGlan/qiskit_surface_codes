```python
import networkx as nx
import matplotlib.pyplot as pyplot
```

```python
def graph_proper(G,edge_label):
    pos=nx.get_node_attributes(G,'pos')
    nx.draw_networkx(G, pos)
    labels = nx.get_edge_attributes(G,edge_label)
    nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
    pyplot.gca().invert_yaxis()
```

# Path Degeneracy Examples

```python
decoder = surface_decoder.GraphDecoder(5,1)
G = decoder.S['Z']
graph_proper(G,'distance')
a = [n for n in G.nodes if n == (0, 1.5, 0.5)][0]
b = [n for n in G.nodes if n == (-1,3.5,4.5)][0]
decoder._path_degeneracy(a, b)
```

```python
matching_graph = decoder.matching_graph(error_graph,'Z')
graph_proper(matching_graph,'weight')
```

```python
g = decoder.matching(matching_graph,'Z')
for e in g:
    print(e)
```
